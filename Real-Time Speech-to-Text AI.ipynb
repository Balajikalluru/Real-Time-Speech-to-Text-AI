```python
import speech_recognition as sr
from langchain_groq import ChatGroq
import pyttsx3

# Function to capture speech and convert it to text
def capture_speech():
    # Initialize recognizer class
    recognizer = sr.Recognizer()

    # Capture input from microphone
    with sr.Microphone() as source:
        print("Adjusting for ambient noise... Please wait.")
        recognizer.adjust_for_ambient_noise(source, duration=1)
        print("Listening for your speech...")

        # Listen to the source
        audio = recognizer.listen(source)

        try:
            # Using Google Web Speech API to recognize the speech
            text = recognizer.recognize_google(audio)
            print("You said: " + text)
            return text
        except sr.UnknownValueError:
            print("Sorry, I couldn't understand the audio.")
        except sr.RequestError as e:
            print(f"Error connecting to Google Web Speech API: {e}")

    return None

# Function to query the ChatGroq model with the converted text
def query_groq_model(text):
    # Initialize the ChatGroq model
    llm = ChatGroq(
        groq_api_key="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",
        model="llama3-8b-8192",
        temperature=0
    )
    
    # Query the model with the converted text
    response = llm.invoke(text)
    return response.content

# Function to convert text to speech
def text_to_speech(text):
    # Initialize the TTS engine
    engine = pyttsx3.init()
    
    # Set properties (optional)
    engine.setProperty('rate', 150)  # Speed of speech
    engine.setProperty('volume', 1)  # Volume level (0.0 to 1.0)

    # Use the engine to say the text
    engine.say(text)
    engine.runAndWait()  # Wait until speech is done

# Main loop for continuous conversation
if __name__ == "__main__":
    while True:
        # Step 1: Capture speech and convert to text
        captured_text = capture_speech()

        # If "exit" is said, break the loop and end the conversation
        if captured_text and captured_text.lower() == "exit":
            print("Exiting conversation...")
            text_to_speech("Goodbye!")
            break

        # Step 2: If speech was successfully captured, query the ChatGroq model
        if captured_text:
            response = query_groq_model(captured_text)
            print("Model response: " + response)
            
            # Step 3: Convert the model's response to speech
            text_to_speech(response)
```
